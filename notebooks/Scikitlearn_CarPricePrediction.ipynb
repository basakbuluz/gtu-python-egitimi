EÄŸitim ve test kÃ¼melerini %80-%20 olacak ÅŸekilde ayÄ±ralÄ±m ve eÄŸitim ve test kÃ¼melerinin boyutlarÄ±nÄ± yazdÄ±ralÄ±m.

**train_test_split** fonksiyonu, veri setini eÄŸitim ve test kÃ¼melerine bÃ¶ler. Bu fonksiyon, eÄŸitim ve test kÃ¼melerinin boyutunu belirlemek iÃ§in **test_size** parametresini kullanÄ±r. AyrÄ±ca, verilerin bÃ¶lÃ¼nmesinin rastgele olmasÄ±nÄ± saÄŸlamak iÃ§in **random_state** parametresini kullanÄ±r.

AÅŸaÄŸÄ±daki kod satÄ±rÄ±, X ve y veri setlerini eÄŸitim ve test kÃ¼melerine bÃ¶ler. EÄŸitim kÃ¼mesi, x_train ve y_train deÄŸiÅŸkenlerine atanÄ±r. Test kÃ¼mesi ise, x_test ve y_test deÄŸiÅŸkenlerine atanÄ±r.

test_size parametresi, test kÃ¼mesinin boyutunu belirtir. Bu parametre, 0 ile 1 arasÄ±nda bir deÄŸer alabilir. Ã–rneÄŸin, test_size=0.2 deÄŸeri, test kÃ¼mesinin boyutunun veri setinin %20'si olacaÄŸÄ±nÄ± belirtir.

random_state parametresi, verilerin bÃ¶lÃ¼nmesinin rastgele olmasÄ±nÄ± saÄŸlar. Bu parametre, bir tamsayÄ± deÄŸeri alabilir. Ã–rneÄŸin, random_state=1 deÄŸeri, verilerin bÃ¶lÃ¼nmesinin her zaman aynÄ± ÅŸekilde yapÄ±lmasÄ±nÄ± saÄŸlar. Bu, deneylerin tekrarlanabilirliÄŸini artÄ±rÄ±r.

EÄŸitim ve test kÃ¼melerine bÃ¶lmek, makine Ã¶ÄŸrenimi modellerinin performansÄ±nÄ± deÄŸerlendirmek iÃ§in Ã¶nemlidir. EÄŸitim kÃ¼mesi, modeli eÄŸitmek iÃ§in kullanÄ±lÄ±r. Test kÃ¼mesi ise, modelin performansÄ±nÄ± deÄŸerlendirmek iÃ§in kullanÄ±lÄ±r.



```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(X, y, test_size=0.2, random_state=1)

print("x train: ", x_train.shape)
print("x test: ", x_test.shape)
print("y train: ", y_train.shape)
print("y test: ", y_test.shape)
```

VarsayÄ±lan (default) hiperparametreler ile LinearRegression modelini tanÄ±mlayalÄ±m ve eÄŸitelim.

DoÄŸrusal regresyon, baÄŸÄ±mlÄ± deÄŸiÅŸken ile bir veya daha fazla baÄŸÄ±msÄ±z deÄŸiÅŸken arasÄ±ndaki doÄŸrusal iliÅŸkiyi modellemek iÃ§in kullanÄ±lan bir makine Ã¶ÄŸrenimi algoritmasÄ±dÄ±r.

x_train ve y_train deÄŸiÅŸkenleri, eÄŸitim verilerini iÃ§erir. EÄŸitim verileri, modelin Ã¶ÄŸrenmesi iÃ§in kullanÄ±lan verilerdir.

lr deÄŸiÅŸkeni, doÄŸrusal regresyon modeli nesnesini iÃ§erir. Bu nesne, eÄŸitim verileriyle eÄŸitildikten sonra yeni veriler Ã¼zerinde tahminler yapmak iÃ§in kullanÄ±labilir.

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(x_train, y_train)
```

Test veri kÃ¼mesi ile tahmin sonuÃ§larÄ±nÄ± elde ediniz.

```python
x_test.iloc[0]
```

```python
predicts=lr.predict(x_test)
predicts
```
